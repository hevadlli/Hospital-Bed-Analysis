```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
import seaborn as sns
```

```python
#Load dataset
hospital_data = "./Data/Hospital_Indonesia_datasets.csv"
df_h = pd.read_csv(hospital_data, delimiter=';')
```

```python
df_h = df_h[['propinsi', 'kab', 'total_tempat_tidur']]
```

```python
df_h.head()
```

```python
bed_per_provinsi = df_h.groupby('propinsi')['total_tempat_tidur'].sum().reset_index()
bed_per_provinsi = bed_per_provinsi.sort_values(by='total_tempat_tidur', ascending=False)
bed_per_provinsi.head()
```

```python
# Visualisasi Bar Chart (Distribusi Tempat Tidur per Provinsi)
plt.figure(figsize=(12, 6))
sns.barplot(data=bed_per_provinsi, x='total_tempat_tidur', y='propinsi', palette="magma")
plt.xlabel("Total Tempat Tidur")
plt.ylabel("Provinsi")
plt.title("Distribusi Tempat Tidur per Provinsi di Indonesia")
plt.show()
```

```python
population_data = "./Data/Populasi_Indonesia.csv"
df_p = pd.read_csv(population_data, delimiter=';')
df_p.head()
```

```python
# Gabungkan kedua dataset berdasarkan provinsi
df_merged = bed_per_provinsi.merge(df_p, left_on="propinsi", right_on="propinsi", how="left")
df_merged['rasio_tempat_tidur'] = (df_merged['total_tempat_tidur'] / df_merged['jumlah'])
```

```python
df_merged.head()
```

```python
# Hitung rata-rata rasio tempat tidur
avg_rasio_tempat_tidur = df_merged['rasio_tempat_tidur'].mean()

# Visualisasi Heatmap (Rasio Tempat Tidur per 1.000 Penduduk)
plt.figure(figsize=(12, 6))
sns.barplot(data=df_merged.sort_values('rasio_tempat_tidur', ascending=False), 
            x='rasio_tempat_tidur', y='propinsi', palette='magma')
plt.axvline(x=avg_rasio_tempat_tidur, color='red', linestyle='--', label="Rata-Rata")
plt.xlabel("Rasio Tempat Tidur per 1.000 Penduduk")
plt.ylabel("Provinsi")
plt.title("Perbandingan Rasio Tempat Tidur Rumah Sakit per Provinsi")
plt.legend()
plt.show()
```

```python
map_indonesia = "./Data/Provinsi.shp"
map_df = gpd.read_file(map_indonesia)
map_df.head()
```

```python
from geopy.geocoders import Nominatim

# Ambil titik tengah (centroid) dari setiap MultiPolygon
map_df['centroid'] = map_df.geometry.centroid

# Inisialisasi geocoder
geolocator = Nominatim(user_agent="geoapi")

# Fungsi untuk mendapatkan nama provinsi berdasarkan titik tengah
def get_province(geometry):
    location = geolocator.reverse((geometry.y, geometry.x), exactly_one=True, language="id")
    if location and 'state' in location.raw['address']:
        return location.raw['address']['state']
    return None
```

```python
# Tambahkan kolom provinsi dengan melakukan geocoding pada centroid
map_df['propinsi'] = map_df['centroid'].apply(get_province)

# Hapus kolom centroid jika sudah tidak dibutuhkan
map_df.drop(columns=['centroid'], inplace=True)

map_df
```

```python
# Ganti 'Daerah Istimewa Yogyakarta' menjadi 'Yogyakarta'
map_df['propinsi'] = map_df['propinsi'].replace('Daerah Istimewa Yogyakarta', 'Yogyakarta')

# Isi nilai None/Nan dengan 'DKI Jakarta'
map_df['propinsi'] = map_df['propinsi'].fillna('DKI Jakarta')

# Cek hasil perubahan
print(map_df['propinsi'].unique())
```

```python
# Gabungkan data dengan peta
map_merged = map_df.merge(df_merged, left_on="propinsi", right_on="propinsi")
map_merged.head()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 8))
map_merged.plot(column='rasio_tempat_tidur', cmap='magma_r', linewidth=0.8, edgecolor='black',
                legend=True, legend_kwds={'label': "Rasio Tempat Tidur per 1.000 Penduduk",
                                          'orientation': "horizontal"}, ax=ax)
ax.set_title("Sebaran Tempat Tidur Rumah Sakit per Provinsi di Indonesia", fontsize=14)
plt.show()
```
